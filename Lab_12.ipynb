# MATRIX CHAIN MULTIPLICATION Dynamic Programming with table filling.
Problem Description:
Find minimum number of scalar multiplications needed to multiply a chain of matrices..

#Theory
Cost of multiplying two matrices:
If A is p×q and B is q×r, cost = p × q × r

Dynamic Programming idea:
Break problem into smaller subproblems.
Store results in a table to avoid recomputation.

m[i][j] = minimum cost to multiply Ai to Aj

## Algorithm
1. Take array p[] of size n
2. Create table m[n][n]
3. Set m[i][i] = 0
4. Fill table diagonally
5. For each chain length L:
    Compute m[i][j]
    Try all possible k between i and j
6. Choose minimum cost
7. Return m[1][n-1]

```python
import sys

def matrix_chain_multiplication(p):
    n = len(p)

    # Create DP table
    m = [[0 for _ in range(n)] for _ in range(n)]

    # L is chain length
    for L in range(2, n):
        for i in range(1, n - L + 1):
            j = i + L - 1
            m[i][j] = sys.maxsize

            for k in range(i, j):
                cost = (m[i][k] +
                        m[k+1][j] +
                        p[i-1] * p[k] * p[j])

                if cost < m[i][j]:
                    m[i][j] = cost


    return m

```

```python
def opt(m,i,j):
    if i==j:
        print(f"A{i}",end="")
    else:
        print("(",end="")
        k=m[i][j]
        opt(m,i,k)
        opt(m,k+1,j)
        print("}",end="")
```

```python
# Input dimensions
p = [10, 30, 5, 60]

table = matrix_chain_multiplication(p)

print("Minimum multiplication cost:", table[1][len(p)-1])

print("\nDP Table:")
for row in table[1:]:
    print(row[1:])
```

```python
def matrix_chain_multi(p):
    n = len(p)

    # Create DP table
    m = [[0 for _ in range(n)] for _ in range(n)]
    K = [[0 for _ in range(n)] for _ in range(n)]
    # L is chain length
    for L in range(2, n):
        for i in range(1, n - L + 1):
            j = i + L - 1
            m[i][j] = sys.maxsize

            for k in range(i, j):
                cost = (m[i][k] +
                        m[k+1][j] +
                        p[i-1] * p[k] * p[j])

                if cost < m[i][j]:
                    m[i][j] = cost
                    K[i][j]=k

    return m,K
```

```python
p = [5, 10, 3, 12, 5, 50, 6]
n=len(p)
table,K=matrix_chain_multi(p)
min_cost=table[1][len(p)-1]
print("Minimum multiplication cost:",min_cost)
total_multi=1
for p in p:
    total_multi*=p
print("Order of multipliction:",end="")
opt(K,1,n-1)
print(end="\n")
print("Saved multiplication =",(total_multi - min_cost))


```
